Creating dataset for Genre recognition using CNN on melspectograms

```python
# songname = "./data/genres/metal/metal.00000.au"
songname = "./data/genres/disco/disco.00000.au"
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import pathlib
```

```python
cmap = plt.get_cmap('inferno')
```

```python
y , sr = librosa.load(songname, mono=True, duration=30)
spectogram = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, n_fft=2048, hop_length=1024)
spectogram = librosa.power_to_db(spectogram, ref=np.max)
plt.axes().set_aspect('equal')
plt.imshow(spectogram, origin="lower", cmap=cmap);
plt.axis('off');
```

Earlier I tried using the above images for the dataset. But the pretrained models are trained on ImageNet and expect square images so squeezed the images as below.

```python
plt.figure(figsize=(10,10))
plt.specgram(y, NFFT=2048, Fs=2, Fc=0, noverlap=128, cmap=cmap, sides='default', mode='default', scale='dB');
plt.axis('off');
plt.savefig('demo.png')
```

Loop for creating the dataset

```python
plt.figure(figsize=(10,10))
genres = 'blues classical country disco hiphop jazz metal pop reggae rock'.split()
for g in genres:
    pathlib.Path(f'img_data/{g}').mkdir(parents=True, exist_ok=True)     
    for filename in os.listdir(f'./data/genres/{g}'):
        songname = f'./data/genres/{g}/{filename}'
        y, sr = librosa.load(songname, mono=True, duration=30)
        plt.specgram(y, NFFT=2048, Fs=2, Fc=0, noverlap=128, cmap=cmap, sides='default', mode='default', scale='dB');
        plt.axis('off');
        plt.savefig(f'img_data/{g}/{filename[:-3].replace(".", "")}.png')
        plt.clf()
    print(f'{g} done!')
```

I didn't create any seperate train and validation set. The fastai library does that while creating the ImageDataBunch.
